```python
%matplotlib inline
import matplotlib.pyplot as plt 
import numpy as np 
import nibabel 
```

```python
shp = nibabel.load('flat.nii.gz').get_fdata().shape
```

```python
i,j,k = np.meshgrid(*[ np.arange(d) for d in shp ])
```

```python
i.shape
```

```python
sin = 60 + 6 * (np.sin(i) + np.sin(j) + np.sin(k))
```

```python
plt.imshow(sin[::,30].T, cmap='hot')
plt.savefig
```

```python
nibabel.save(nibabel.Nifti1Image(sin, np.eye(4)), 'sine_field.nii.gz')
flat = 60 * np.ones_like(sin)
nibabel.save(nibabel.Nifti1Image(flat, np.eye(4)), 'flat_field.nii.gz')
```
